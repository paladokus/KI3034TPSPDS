```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the text files into dataframes
# Import data from text files
df_abs = pd.read_csv('DataPoints_Rhodamin B_2ppm_Abs_AbsTSpec (1).txt', sep='\t', header=None, skiprows=12)
df_ems_350nm_em = pd.read_csv('DataPoints_Rhodamin B_2ppm_Ems_350nm_EM.txt', sep='\t', header=None, skiprows=12)
df_ems_553nm_em = pd.read_csv('DataPoints_Rhodamin B_2ppm_Ems_350nm_EM.txt', sep='\t', header=None, skiprows=12)

# Rename columns (only keep the first two columns and rename them)
df_abs = df_abs.iloc[:, :2]
df_abs.columns = ['Wavelength (nm)', 'Absorbance (a.u.)']

df_ems_350nm_em = df_ems_350nm_em.iloc[:, :2]
df_ems_350nm_em.columns = ['Wavelength (nm)', 'Intensity (a.u.)']

df_ems_553nm_em = df_ems_553nm_em.iloc[:, :2]
df_ems_553nm_em.columns = ['Wavelength (nm)', 'Intensity (a.u.)']

# Display the first few rows of each dataframe
print("Absorption Data:")
print(df_abs.head())

print("\nEmission Data (350nm):")
print(df_ems_350nm_em.head())

print("\nEmission Data (553nm):")
print(df_ems_553nm_em.head())
```

```python
import matplotlib.pyplot as plt

# Plot df_abs
plt.figure(figsize=(10, 6))
plt.plot(df_abs['Wavelength (nm)'], df_abs['Absorbance (a.u.)'], label='Absorbance', color='blue')
plt.title('Absorbance vs Wavelength')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance (a.u.)')
plt.legend()
plt.grid(True)
plt.show()

# Plot df_ems_350nm_em
plt.figure(figsize=(10, 6))
plt.plot(df_ems_350nm_em['Wavelength (nm)'], df_ems_350nm_em['Intensity (a.u.)'], label='Emission (350nm)', color='green')
plt.title('Emission Intensity vs Wavelength (350nm)')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (a.u.)')
plt.legend()
plt.grid(True)
plt.show()

# Plot df_ems_553nm_em
plt.figure(figsize=(10, 6))
plt.plot(df_ems_553nm_em['Wavelength (nm)'], df_ems_553nm_em['Intensity (a.u.)'], label='Emission (553nm)', color='red')
plt.title('Emission Intensity vs Wavelength (553nm)')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity (a.u.)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Normalize the dataframes and add as new columns
df_abs['Normalized Absorbance'] = (df_abs['Absorbance (a.u.)'] / df_abs['Absorbance (a.u.)'].max()) * 100
print(df_abs)
df_ems_350nm_em['Normalized Intensity'] = (df_ems_350nm_em['Intensity (a.u.)'] / df_ems_350nm_em['Intensity (a.u.)'].max()) * 100
print(df_ems_350nm_em)
df_ems_553nm_em['Normalized Intensity'] = (df_ems_553nm_em['Intensity (a.u.)'] / df_ems_553nm_em['Intensity (a.u.)'].max()) * 100
print(df_ems_553nm_em)
```

```python
# Plot real data
plt.figure(figsize=(12, 6))
plt.plot(df_abs['Wavelength (nm)'], df_abs['Absorbance (a.u.)'], label='Absorbance', color='blue')
plt.plot(df_ems_350nm_em['Wavelength (nm)'], df_ems_350nm_em['Intensity (a.u.)'], label='Emission (350nm)', color='green')
plt.plot(df_ems_553nm_em['Wavelength (nm)'], df_ems_553nm_em['Intensity (a.u.)'], label='Emission (553nm)', color='red')
plt.title('Real Data: Absorbance and Intensity vs Wavelength')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Value (a.u.)')
plt.legend()
plt.grid(True)
plt.show()

# Plot normalized data for Absorbance and Emission (350nm)
plt.figure(figsize=(12, 6))
plt.plot(df_abs['Wavelength (nm)'], df_abs['Normalized Absorbance'], label='Normalized Absorbance', color='blue')
plt.plot(df_ems_350nm_em['Wavelength (nm)'], df_ems_350nm_em['Normalized Intensity'], label='Normalized Emission (350nm)', color='green', alpha=0.7)
plt.title('Normalized Data: Absorbance and Emission (350nm) vs Wavelength')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Value (%)')
plt.legend()
plt.grid(True)
plt.show()

# Plot normalized data for Absorbance and Emission (553nm)
plt.figure(figsize=(12, 6))
plt.plot(df_abs['Wavelength (nm)'], df_abs['Normalized Absorbance'], label='Normalized Absorbance', color='blue')
plt.plot(df_ems_553nm_em['Wavelength (nm)'], df_ems_553nm_em['Normalized Intensity'], label='Normalized Emission (553nm)', color='red', alpha=0.7)
plt.title('Normalized Data: Absorbance and Emission (553nm) vs Wavelength')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Value (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Find the peak wavelength for absorbance
peak_wavelength_abs = df_abs.loc[df_abs['Absorbance (a.u.)'].idxmax(), 'Wavelength (nm)']
print(f"Peak Wavelength for Absorbance: {peak_wavelength_abs} nm")

# Find the peak wavelength for emission (350nm)
peak_wavelength_em_350nm = df_ems_350nm_em.loc[df_ems_350nm_em['Intensity (a.u.)'].idxmax(), 'Wavelength (nm)']
print(f"Peak Wavelength for Emission (350nm): {peak_wavelength_em_350nm} nm")

# Find the peak wavelength for emission (553nm)
peak_wavelength_em_553nm = df_ems_553nm_em.loc[df_ems_553nm_em['Intensity (a.u.)'].idxmax(), 'Wavelength (nm)']
print(f"Peak Wavelength for Emission (553nm): {peak_wavelength_em_553nm} nm")
```

```python
# Plot normalized absorbance vs normalized emission (350nm) with peak label
plt.figure(figsize=(12, 6))
plt.plot(df_abs['Wavelength (nm)'], df_abs['Normalized Absorbance'], label='Normalized Absorbance', color='blue')
plt.plot(df_ems_350nm_em['Wavelength (nm)'], df_ems_350nm_em['Normalized Intensity'], label='Normalized Emission (350nm)', color='green', alpha=0.7)
plt.scatter(peak_wavelength_abs, df_abs.loc[df_abs['Wavelength (nm)'] == peak_wavelength_abs, 'Normalized Absorbance'], color='red', label='Abs Peak')
plt.scatter(peak_wavelength_em_350nm, df_ems_350nm_em.loc[df_ems_350nm_em['Wavelength (nm)'] == peak_wavelength_em_350nm, 'Normalized Intensity'], color='orange', label='Em Peak (350nm)')
plt.annotate(f'Abs Peak: {peak_wavelength_abs} nm', 
             xy=(peak_wavelength_abs, df_abs.loc[df_abs['Wavelength (nm)'] == peak_wavelength_abs, 'Normalized Absorbance'].values[0]), 
             xytext=(peak_wavelength_abs - 70, df_abs['Normalized Absorbance'].max() * 0.8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
# Annotate the peak for Emission (350nm)
plt.annotate(f'Em Peak (350nm): {peak_wavelength_em_350nm:.2f} nm', 
             xy=(peak_wavelength_em_350nm, df_ems_350nm_em.loc[df_ems_350nm_em['Wavelength (nm)'] == peak_wavelength_em_350nm, 'Normalized Intensity'].values[0]), 
             xytext=(peak_wavelength_em_350nm + 20, df_ems_350nm_em['Normalized Intensity'].max() * 0.8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title('Normalized Absorbance vs Normalized Emission (350nm) with Peak Labels')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Value (%)')
plt.legend()
plt.grid(True)  # Add grid to the plot
plt.show()

# Plot normalized absorbance vs normalized emission (553nm) with peak label
plt.figure(figsize=(12, 6))
plt.plot(df_abs['Wavelength (nm)'], df_abs['Normalized Absorbance'], label='Normalized Absorbance', color='blue')
plt.plot(df_ems_553nm_em['Wavelength (nm)'], df_ems_553nm_em['Normalized Intensity'], label='Normalized Emission (553nm)', color='red', alpha=0.7)
plt.scatter(peak_wavelength_abs, df_abs.loc[df_abs['Wavelength (nm)'] == peak_wavelength_abs, 'Normalized Absorbance'], color='red', label='Abs Peak')
plt.scatter(peak_wavelength_em_553nm, df_ems_553nm_em.loc[df_ems_553nm_em['Wavelength (nm)'] == peak_wavelength_em_553nm, 'Normalized Intensity'], color='orange', label='Em Peak (553nm)')
plt.annotate(f'Abs Peak: {peak_wavelength_abs} nm', 
             xy=(peak_wavelength_abs, df_abs.loc[df_abs['Wavelength (nm)'] == peak_wavelength_abs, 'Normalized Absorbance'].values[0]), 
             xytext=(peak_wavelength_abs - 70, df_abs['Normalized Absorbance'].max() * 0.8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Em Peak (553nm): {peak_wavelength_em_553nm:.2f} nm', 
             xy=(peak_wavelength_em_553nm, df_ems_553nm_em.loc[df_ems_553nm_em['Wavelength (nm)'] == peak_wavelength_em_553nm, 'Normalized Intensity'].values[0]), 
             xytext=(peak_wavelength_em_553nm + 20, df_ems_553nm_em['Normalized Intensity'].max() * 0.8),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.title('Normalized Absorbance vs Normalized Emission (553nm) with Peak Labels')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Value (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate the Stokes shift
stokes_shift_350nm = peak_wavelength_em_350nm - peak_wavelength_abs
print(f"Stokes Shift (350nm Emission): {stokes_shift_350nm} nm")

# Repeat for other emission data (e.g., 553nm)
peak_wavelength_em_553nm = df_ems_553nm_em.loc[df_ems_553nm_em['Intensity (a.u.)'].idxmax(), 'Wavelength (nm)']
stokes_shift_553nm = peak_wavelength_em_553nm - peak_wavelength_abs
print(f"Stokes Shift (553nm Emission): {stokes_shift_553nm} nm")

# Calculate energy corresponding to the wavelengths
h = 6.626e-34  # Planck's constant in J·s
c = 3.0e8      # Speed of light in m/s

# Convert wavelengths from nm to meters
lambda_abs_m = peak_wavelength_abs * 1e-9
lambda_em_350nm_m = peak_wavelength_em_350nm * 1e-9
lambda_em_553nm_m = peak_wavelength_em_553nm * 1e-9

# Calculate energy in Joules
energy_abs = h * c / lambda_abs_m
energy_em_350nm = h * c / lambda_em_350nm_m
energy_em_553nm = h * c / lambda_em_553nm_m

print(f"Energy for Absorbance Peak: {energy_abs:.3e} J")
print(f"Energy for Emission Peak (350nm): {energy_em_350nm:.3e} J")
print(f"Energy for Emission Peak (553nm): {energy_em_553nm:.3e} J")
```
